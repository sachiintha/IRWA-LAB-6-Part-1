#3.	Python implementation to print first <h1> tag from https://en.wikipedia.org/wiki/Main_Page      

```python
#get the header of a web page 

from urllib.request import urlopen

from bs4 import BeautifulSoup 

html = urlopen("https://en.wikipedia.org/wiki/Main_Page")

bs = BeautifulSoup(html) #to scrap the data from source code 

titles = bs.find_all({'h1'})

print(' list the first header tag :', *titles, sep='\n\n')
```

#4. How to print all the header tags from en.wikipedia.org/wiki/Main_Page.

```python
#get the header of a web page 

from urllib.request import urlopen

from bs4 import BeautifulSoup 

html = urlopen("https://en.wikipedia.org/wiki/Main_Page")

bs = BeautifulSoup(html) #to scrap the data from source code 

titles = bs.find_all({'h1','h2','h3'})
print(' list the first header tag :', titles, sep='\n\n')

print(' list the first header tag :', *titles, sep='\n\n')

# * lists the all the header tags 
```

#5. How to display all the image links from https://www.ibm.com/lk-en.

```python
#get the header of a web page 

#relative URLs and absolute URLs

#img write to select image tags   
from urllib.request import urlopen
from bs4 import BeautifulSoup 
import re 

html = urlopen("https://www.ibm.com/us-en/")
bs = BeautifulSoup(html) #to scrap the data from source code 

images = bs.find_all({'img','html.parser'})

for image in images:
  print(image['src']+'/n')

  print(image)
# * lists the all the header tags 
```

#6. Python implementation to retrieve all the links in a particular URL.

```python
# display tags links --> href 

from urllib.request import urlopen
from bs4 import BeautifulSoup 
import re 

html = urlopen("https://www.sliit.lk/")
bs = BeautifulSoup(html) #to scrap the data from source code 

links = bs.find_all('a')   #every href tag starts with a 

for link in links:
  print(link.get('href'))

  print(link.text)
```

#7. Python implementation to display section that contain a specified string in a given web page.

```python
#how to exract puricular space whic extract ehe link 

import requests

from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Main_Page"

req = requests.get(url)

bs = BeautifulSoup(req.text, 'lxml')

str1 = bs.find_all(string=re.compile('Hurricane')) #one chracter set can't use spaces 

for txt in str1:
  print(txt.split()) 
```

#8. Python implementation to obtain html code of parent content of a particular html tag.

```python
#child inside one tag there are several tags 
import requests
from bs4 import BeautifulSoup 
 

url = ("https://www.sliit.lk/")

req = requests.get(url)

bs = BeautifulSoup(req.text, 'lxml')

print(bs.title.text)

print(bs.title.parent)
```

#9. Python implementation to obtain html code of child nodes inside a particular html tag
